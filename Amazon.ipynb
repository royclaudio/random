```python
def minimalHeaviestSetA(arr):
    A = []
    B = []
    arr.sort(reverse=True)
    A , B, arr = helper(arr, A , B)
    return A

def helper(arr, A ,B):
    if(sum(A) == 0):
        A.append(arr[0])
        A.append(arr[1])
        arr.remove(arr[0])
        arr.remove(arr[1])
        B = arr
        helper(arr, A,B)
    if(sum(A) > sum(B)):
        A.sort()
        return A, B, arr 
    else:
        A.append(arr[0])
        arr.remove(arr[0])
        B = arr
        helper(arr, A, B)
    return A,B, arr
```

```python
def findcircle(isconnected):
    visited = [False for i in range(len(isconnected))]
    res = 0
    for i in range(len(visited)):
        if(visited[i] == False):
            dfs(isconnected, i, visited)
            res+=1
    return res

def dfs(isconnected, i, visited):
    if(visited[i] == True):
        return
    visited[i] = True
    for j in range(len(isconnected)):
        if(isconnected[i][j] == "1" and j!=i):
            dfs(isconnected,j,visited)

def printMatrix(A):
    # Loop to iterate over the matrix
    for i in range(len(A)):
        for j in range(len(A[0])):
            print(A[i][j], end = ' ')
        print()
```

```python
def twoSum(nums, target) -> list[int]:
        comps = {}
        for i, num in enumerate(nums):
            potential_match = target - num
            if potential_match in comps:
                return [comps[potential_match], i]
            else:
                comps[num] = i
        return [-1,-1]
```

```python
arr = [2,7,11,15]
target = 9
twosum(arr,target)
```

```python
import heapq

def findMedianSortedArrays(nums1,nums2) -> float:
    final = heapq.merge(nums1,nums2)
    final = list(final)
    if len(final)%2 != 0:
        return final[len(final)//2]
    else:
        left = int(len(final)/2) - 1
        right = int(len(final)/2)
        return (final[left]+final[right])/2
```

```python
nums1 = [1,3]
nums2 = [2]
findMedianSortedArrays(nums1,nums2)
```

```python
def longestPalindrome(s) -> str:
    if s == s[::-1]: return s

    start, size = 0, 1
    for i in range(1, len(s)):
        l, r = i - size, i + 1
        s1, s2 = s[l-1:r], s[l:r]

        if l - 1 >= 0 and s1 == s1[::-1]:
            start, size = l - 1, size + 2
        elif s2 == s2[::-1]:
            start, size = l, size + 1

    return s[start:start+size]
```

```python
string = "babad"
longestPalindrome(string)
```

```python
import re
def myAtoi(string):
        """
        :type str: str
        :rtype: int
        """
        match = re.match(r'^\s*([+|-]?\d+)', string)
        
        if match:
            integer = int(match.group())
            return max(-(1 << 31), min(integer, (1 << 31) - 1))
        
        return 0
```

```python
string = "4193 with words"
myAtoi(string)
```

```python
def combinationSum(candidates, target):
    ret = []
    dfs(candidates, target, [], ret)
    return ret

def dfs(nums, target, path, ret):
    if target < 0:
        return 
    if target == 0:
        ret.append(path)
        return 
    for i in range(len(nums)):
        dfs(nums[i:], target-nums[i], path+[nums[i]], ret)
```

```python
candidates = [2,3,6,7]
target = 7
combinationSum(candidates, target)
```

```python
def minPathSum(grid) -> int:
    
        n = len(grid)
        m = len(grid[0])
       
        # Edit the first row:
        for i in range(1, m):
            grid[0][i] = grid[0][i-1] + grid[0][i]
        
        # Edit the first col:
        for i in range(1, n):
            grid[i][0] = grid[i-1][0] + grid[i][0]
        
        # Edit the remaining values in the grid accordingly!
        for i in range(1, n):
            for j in range(1, m):
                grid[i][j] = min(grid[i-1][j], grid[i][j-1]) + grid[i][j]
        
        
        return grid[-1][-1]
```

```python
grid = [[1,3,1],[1,5,1],[4,2,1]]
minPathSum(grid)
```
